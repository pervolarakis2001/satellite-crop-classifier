```python
import zipfile
import os

def unzip_to_dataset(zip_path, output_folder="dataset"):
    # Create output directory if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Unzip the file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(output_folder)
        print(f"Extracted '{zip_path}' to '{output_folder}'")

# Example usage
zip_file_path = '/home/ubuntu/bdeo_exercise_2/timematch_data.zip'  # Replace with your actual zip file path
unzip_to_dataset(zip_file_path)
```
